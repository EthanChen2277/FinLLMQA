```python
import requests
from core import SERVER_API_URL, SERVER_LLM_API_PORT

def embedding_request(text: str):
    body = {
        'input': text
    }
    res = requests.post(SERVER_API_URL + SERVER_LLM_API_PORT + '/v1/embeddings', json=body)
    if res.status_code == '200':
        data = res.json()['data']
        return data
    else:
        return None
```

```python
text = '贵州茅台'
body = {
        'input': text
    }
res = requests.post(SERVER_API_URL + SERVER_LLM_API_PORT + '/v1/embeddings', json=body)
```
